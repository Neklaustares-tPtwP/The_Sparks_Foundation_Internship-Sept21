## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

## Reading the Dataset

```python
df = pd.read_csv("Iris.csv")
df.head()
```

## Data Preprocessing

```python
# Dropping the 'Id' column as it is not important
df = df.iloc[:,1:]
df.head()
```

## Data Visualization

```python
sns.set_style('darkgrid')
sns.pairplot(df, hue='Species', height=4)
plt.show()
```

## Training the Decision Tree Classifier

```python
# Splitting data into training and testing data
X = df.drop('Species',axis=1)
y = df.Species

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19, stratify=y)
```

```python
# Creating Decision Tree Classifier and training it on train data
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

print("Decision Tree Classifier Trained!!")
```

## Visualizing the Decision Tree graphically

```python
# List of different features and classes
features = list(df.columns[:-1])
classes = list(df.Species.unique())

# Setting dpi = 300 to make image clearer than default
fig, axes = plt.subplots(figsize = (4,4), dpi=300)

# Visualizing decision tree graphically
tree.plot_tree(dt_classifier,
           feature_names = features, 
           class_names=classes,
           filled=True, rounded=True, impurity=True);
```
